### 학습목표 
 1. requests 모듈 사용하여 http request/response 확인하기

#### requests 모듈
1. http request/response를 위한 모듈
2. HTTP method를 메소드 명으로 사용하여 request 요청 예) get, post

```python
import requests #임폴트 먼저
```

#### get 요청하기
1. http get 요청하기
2. query parameter 이용하여 데이터 전달하기

```python
url = 'https://news.v.daum.net/v/20190728165812603'
resp = requests.get(url)

# resp #200의 의미는 문제가 없다는 뜻이다!!
resp.text
```

#### post 요청하기
1. http post 요청하기
2. post data 이용하여 데이터 전달하기
> 보통 로그인 정보이다

```python
# post는 data가 필요하다!!
url = 'https://www.kangcom.com/member/member_check.asp'
data = {
    'id':'macmath22',
    'pwd':'Test1357!'
}
resp = requests.post(url,data=data)
resp.text
```

```python
# 잘못된 비밀번호를 입력했을 시
url = 'https://www.kangcom.com/member/member_check.asp'
data = {
    'id':'macmath22',
    'pwd':'T!'
}
resp = requests.post(url,data=data)
resp.text
```

#### HTTP header 데이터 이용하기
1. header 데이터 구성하기
2. header 데이터 전달하기

```python
url = 'https://news.v.daum.net/v/20190728165812603'
headers = {
    'user-agent':'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/81.0.4044.138 Safari/537.36'
    
}

resp = requests.get(url, headers=headers)

resp.text
```

#### HTTP response 처리하기
1. response 객체의 이해
2. status_code 확인하기
3. text 속성 확인하기

```python
url = 'https://news.v.daum.net/v/20190728165812603'
resp = requests.get(url)

if resp.status_code == 200:
    print(resp.text)
else:
    print('error')
```
